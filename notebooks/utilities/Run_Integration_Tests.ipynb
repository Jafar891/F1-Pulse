```python
# =============================================================================
# F1-Pulse | Integration Test Suite Runner
# Notebook: 00_Run_Integration_Tests
# Author:   Jafar891
# Updated:  2026
#
# Executes all integration tests for the F1-Pulse pipeline on Databricks Serverless.
# Tests verify Bronze → Silver → Gold transformations end-to-end.
#
# Tests must be run via pytest.main() in a notebook Python cell — NOT via %sh.
# The notebook cell has access to the active Spark Connect session; %sh subprocesses do not.
#
# Usage:
#   Run this notebook cell directly. All tests in tests/integration_tests/ will execute.
#   Exit code 0 = all tests passed.
#
# Idempotent — safe to re-run.
# =============================================================================

# =============================================================================
# F1-Pulse | Integration Test Suite Runner
# Notebook: Run_Integration_Tests
# Author:   Jafar891
# Updated:  2026
#
# Executes all integration tests for the F1-Pulse pipeline on Databricks Serverless.
# Tests verify Bronze → Silver → Gold transformations end-to-end.
#
# Tests must be run via pytest.main() in a notebook Python cell — NOT via %sh.
# The notebook cell has access to the active Spark Connect session; %sh subprocesses do not.
#
# Usage:
#   Run this notebook cell directly. All tests in tests/integration_tests/ will execute.
#   Exit code 0 = all tests passed.
#
# Idempotent — safe to re-run.
# =============================================================================

import os
import sys

# 1. Double-lock to prevent bytecode writing BEFORE any imports
os.environ["PYTHONDONTWRITEBYTECODE"] = "1"
sys.dont_write_bytecode = True

# Now it is safe to import pytest
import pytest

# 2. Define the path
repo_path = "/Workspace/Users/jafar.gahramanov@gmail.com/F1-Pulse"
if repo_path not in sys.path:
    sys.path.insert(0, repo_path)

# 3. Run integration tests with maximum cache-busting flags
exit_code = pytest.main([
    f"{repo_path}/tests/integration_tests/",
    "-v",
    "--no-header",
    "-p", "no:cacheprovider",   # Disables .pytest_cache
    "--keep-duplicates"         # Sometimes helps with Databricks FUSE imports
])

if exit_code == 0:
    print("\n✅ INTEGRATION TEST SUITE PASSED SUCCESSFULLY")
else:
    print(f"\n❌ INTEGRATION TEST SUITE FAILED with code: {exit_code}")
```
