```python
import os
import sys

# 1. Double-lock to prevent bytecode writing BEFORE any imports
os.environ["PYTHONDONTWRITEBYTECODE"] = "1"
sys.dont_write_bytecode = True

# Now it is safe to import pytest
import pytest

# 2. Define the path
repo_path = "/Workspace/Users/jafar.gahramanov@gmail.com/F1-Pulse"
if repo_path not in sys.path:
    sys.path.insert(0, repo_path)

# 3. Run tests with maximum cache-busting flags
exit_code = pytest.main([
    f"{repo_path}/tests/unit_tests/",
    "-v",
    "--no-header",
    "-p", "no:cacheprovider",   # Disables .pytest_cache
    "--keep-duplicates"         # Sometimes helps with Databricks FUSE imports
])

if exit_code == 0:
    print("\n✅ TEST SUITE PASSED SUCCESSFULLY")
else:
    print(f"\n❌ TEST SUITE FAILED with code: {exit_code}")
```
