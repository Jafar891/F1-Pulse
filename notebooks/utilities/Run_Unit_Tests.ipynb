```python
import pytest
import sys
import os

# 1. Define your project path
repo_path = "/Workspace/Users/jafar.gahramanov@gmail.com/F1-Pulse"

# 2. Change the cluster's current working directory
os.chdir(repo_path)

# 3. Add to system path
if repo_path not in sys.path:
    sys.path.insert(0, repo_path) 

# ==========================================
# THE FIX: Disable __pycache__ creation
# ==========================================
sys.dont_write_bytecode = True

print("🚀 Running Unit Tests...\n" + "="*40)

# Added the "-p no:cacheprovider" flag to stop Pytest from making cache folders
retcode = pytest.main([
    "-v", 
    "-p", "no:cacheprovider", 
    "tests/"
])

if retcode == 0:
    print("\n✅ ALL TESTS PASSED!")
else:
    raise RuntimeError("❌ Pytest failed! Check the logs above.")
```
